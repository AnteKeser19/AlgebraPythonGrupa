```python
import random

def baci_kockicu(br_str=6,br_bacanja = 1, br_kockica = 1):
    for kockica in range(br_kockica):
        for bacanje in range(br_bacanja):
            rezultat = random.randrange(1,br_str+1)
            print(rezultat,end=' ')
        print()
    print()
```

```python
def suma_kockica(br_bacanja = 1, br_kockica = 1):
    suma = 0
    for kockica in range(br_kockica):
        for bacanje in range(br_bacanja):
            rezultat = random.randrange(1,7)
            suma += rezultat
        return suma
```

```python
import re

gradovi = ['     Zagreb    ', 'Osijek!', 'Rijeka', 'rijeka', 'PUla', 'slavonski brod##', 'Nova gradiska?','Grad123']

def popravi_unos(lista):
    rezultat = []
    for element in lista:
        element = element.strip()
        element = re.sub('[!#?]','',element)
        element = re.sub('[0-9]','',element)
        element = element.capitalize()
        rezultat.append(element)
    return rezultat

popravi_unos(gradovi)
```

```python
def temp_to_fh(lista):
    konvertirano = []
    for element in lista:
        element = 9/5 * element + 32
        element = round(element, 2) 
        konvertirano.append(element)
    print(konvertirano)

temp_to_fh(dnevne_temperature)
```

```python
class Djelatnik():
    def __init__(self,placa,ime,prezime):
        self.placa = placa
        self.ime = ime
        self.prezime = prezime
        self.prirez = 0.00

    def izracunaj_prirez(self,iznos_prireza):
        self.prirez = self.placa * iznos_prireza

programer = Djelatnik(5000.0,'Mate','Matic')
print('ime/prezime',programer.ime,programer.prezime)
print('placa',programer.placa)
programer.izracunaj_prirez(0.12)
print('prirez',programer.prirez)
```
